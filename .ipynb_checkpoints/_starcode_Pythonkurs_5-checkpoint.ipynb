## Mehrere Linien in einem Diagramm

Manchmal möchten wir mehrere Linien in einem Diagramm darstellen, um Daten zu vergleichen. Zum Beispiel, vergleichen wir die Anzahl der entdeckten Sterne mit der Anzahl der entdeckten Planeten:

```python
sterne_entdeckt = [3, 5, 7, 9, 10, 11, 15, 17, 20, 22, 25, 28]
planeten_entdeckt = [1, 2, 4, 4, 5, 6, 8, 9, 10, 11, 12, 13]

plt.plot(monate, sterne_entdeckt, label='Sterne')
plt.plot(monate, planeten_entdeckt, label='Planeten')
plt.title('Entdeckte Sterne und Planeten pro Monat')
plt.xlabel('Monate')
plt.ylabel('Anzahl der entdeckten Objekte')
plt.legend()
plt.show()
```

## Balkendiagramme

Manchmal möchten wir statt Liniendiagrammen Balkendiagramme verwenden, um Daten zu vergleichen. Zum Beispiel können wir die Anzahl der entdeckten Sterne und Planeten pro Monat in einem Balkendiagramm darstellen:

```python
breite = 0.35
monate_balken = monate - breite / 2

plt.bar(monate_balken, sterne_entdeckt, width=breite, label='Sterne')
plt.bar(monate_balken + breite, planeten_entdeckt, width=breite, label='Planeten')
plt.title('Entdeckte Sterne und Planeten pro Monat')
plt.xlabel('Monate')
plt.ylabel('Anzahl der entdeckten Objekte')
plt.legend()
plt.show()
```

## Punktediagramme (Scatterplots)

Ein weiteres nützliches Diagramm ist der Scatterplot, bei dem Punkte anstelle von Linien oder Balken verwendet werden, um Daten darzustellen. Zum Beispiel können wir die Größe der entdeckten Sterne im Vergleich zu ihrer Helligkeit darstellen:



```python
sterne_groesse = np.random.rand(50) * 10
sterne_helligkeit = np.random.rand(50) * 100

plt.scatter(sterne_groesse, sterne_helligkeit)
plt.title('Sterngröße im Vergleich zur Helligkeit')
plt.xlabel('Größe (in Sonnenradien)')
plt.ylabel('Helligkeit (in Lumen)')
plt.show()
```

## Histogramme

Histogramme helfen uns, die Verteilung von Daten zu visualisieren. Zum Beispiel können wir die Verteilung der Entfernungen von Planeten zu ihrem Stern darstellen:



```python
# array mit der Entfernung von 100 Planeten
planeten_entfernungen = np.random.normal(5, 2, 100)
print(planeten_entfernungen)
```

```python
plt.hist(planeten_entfernungen, bins=10)
plt.title('Verteilung der Entfernungen von Planeten zu ihrem Stern')
plt.xlabel('Entfernung (in Astronomischen Einheiten)')
plt.ylabel('Anzahl der Planeten')
plt.show()
```

## Subplots

Manchmal möchten wir mehrere Diagramme nebeneinander anzeigen, um sie leichter vergleichen zu können. Zum Beispiel können wir die Größe und Helligkeit von Sternen und Planeten in zwei verschiedenen Scatterplots darstellen:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

sterne_groesse = np.random.rand(50) * 10
sterne_helligkeit = np.random.rand(50) * 100
planeten_groesse = np.random.rand(50) * 2
planeten_helligkeit = np.random.rand(50) * 10

ax1.scatter(sterne_groesse, sterne_helligkeit)
ax1.set_title('Sterngröße im Vergleich zur Helligkeit')
ax1.set_xlabel('Größe (in Sonnenradien)')
ax1.set_ylabel('Helligkeit (in Lumen)')

ax2.scatter(planeten_groesse, planeten_helligkeit)
ax2.set_title('Planetengröße im Vergleich zur Helligkeit')
ax2.set_xlabel('Größe (in Erdradien)')
ax2.set_ylabel('Helligkeit (in Lumen)')

plt.show()
```

## 3D-Diagramme

Mit Matplotlib können wir auch 3D-Diagramme erstellen. Zum Beispiel können wir die Positionen von Sternen im Raum in einem 3D-Scatterplot darstellen:

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50) * 100
y = np.random.rand(50) * 100
z = np.random.rand(50) * 100

ax.scatter(x, y, z)
ax.set_title('Positionen von Sternen im Raum')
ax.set_xlabel('X-Koordinate (in Lichtjahren)')
ax.set_ylabel('Y-Koordinate (in Lichtjahren)')
ax.set_zlabel('Z-Koordinate (in Lichtjahren)')

plt.show()
```

## Anpassen der Diagrammstile

Matplotlib bietet verschiedene vordefinierte Stile, mit denen wir das Aussehen unserer Diagramme schnell ändern können. Um die verfügbaren Stile anzuzeigen, können wir plt.style.available verwenden:

```python
print(plt.style.available)
```

Um einen bestimmten Stil zu verwenden, können wir plt.style.use() aufrufen:

```python
plt.style.use('seaborn-darkgrid')

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sinusfunktion')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```
